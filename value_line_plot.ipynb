```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.cm as cm
```

```python
eda_plot_data = pd.read_csv('data/KEMDy20_v1_1/eda.csv')
ibi_plot_data = pd.read_csv('data/KEMDy20_v1_1/ibi.csv')
temp_plot_data = pd.read_csv('data/KEMDy20_v1_1/temp.csv')
```

```python
sample_eda = pd.DataFrame(columns=temp_plot_data.columns)
for g in temp_plot_data['Emotion'].unique():
    temp_df = temp_plot_data.query('Emotion==@g').sample(n=20) ## 그룹별 데이터 추출 및 2개 비복원 추출
    sample_eda = pd.concat([sample_eda,temp_df]) ## 데이터 추가
sample_eda = sample_eda.reset_index(drop=True) ## 인덱스 초기화

# create a dictionary to map old column names to new ones
column_mapping = {f'temp{i}': str(i) for i in range(1, 142)}

# rename the columns using the dictionary
sample_eda.rename(columns=column_mapping, inplace=True)
```

```python
sample_ibi = pd.DataFrame(columns=temp_plot_data.columns)
for g in temp_plot_data['Emotion'].unique():
    temp_df = temp_plot_data.query('Emotion==@g').sample(n=20) ## 그룹별 데이터 추출 및 2개 비복원 추출
    sample_ibi = pd.concat([sample_ibi,temp_df]) ## 데이터 추가
sample_ibi = sample_ibi.reset_index(drop=True) ## 인덱스 초기화

# create a dictionary to map old column names to new ones
column_mapping = {f'temp{i}': str(i) for i in range(1, 142)}

# rename the columns using the dictionary
sample_ibi.rename(columns=column_mapping, inplace=True)
```

```python
sample_temp = pd.DataFrame(columns=temp_plot_data.columns)
for g in temp_plot_data['Emotion'].unique():
    temp_df = temp_plot_data.query('Emotion==@g').sample(n=20) ## 그룹별 데이터 추출 및 2개 비복원 추출
    sample_temp = pd.concat([sample_temp,temp_df]) ## 데이터 추가
sample_temp = sample_temp.reset_index(drop=True) ## 인덱스 초기화

# create a dictionary to map old column names to new ones
column_mapping = {f'temp{i}': str(i) for i in range(1, 142)}

# rename the columns using the dictionary
sample_temp.rename(columns=column_mapping, inplace=True)
```

```python
# Create a list of labels
labels=['angry', 'happy', 'neutral', 'sad', 'surprise']

# Create a list of patches for each label
colors = cm.brg(np.linspace(0, 1, len(labels)))
patches = [mpatches.Patch(color=color, label=label) for label, color in zip(labels, colors)]

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot the line plot for each row of the DataFrame with the corresponding color
for index, row in sample_eda.replace(0, np.nan).iterrows():
    plt.plot(row[5:], label=row['Emotion'], color=colors[labels.index(row['Emotion'])])

# Set the title, x-label, y-label, and legend
plt.title('EDA Line Plot')
plt.xlabel('time')
plt.ylabel('Value')
plt.legend(handles=patches, loc='upper right', title='Emotion')
plt.style.use('seaborn-white')

# Set the x-axis ticks to show every 10
plt.xticks(np.arange(0, len(sample_eda.columns[5:]), 10))

# Show the plot
plt.show()
```
