```python
import os
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
directory= 'C:/Users/user1/Desktop/공부자료/제2회 ETRI 휴먼이해 인공지능 논문경진대회/data/KEMDy20_v1_1/'
os.chdir(directory)
```

```python
eda = os.path.join(directory, 'EDA/')
files_eda = os.listdir(eda)
filescnt_eda = [e for e in files_eda]
df_eda = pd.DataFrame()

for e in filescnt_eda :
    ann = os.path.join(directory, 'EDA/', e)
    files = os.listdir(ann)
    filescnt_wav = [i for i in files if i.endswith('.csv')]
    df_all = pd.DataFrame()
    
    for i in range(0, len(filescnt_wav)) :
        file = filescnt_wav[i]
        df = pd.read_csv(os.path.join(ann, file), names=['EDA', 'time', 'id']) 
        df_all = pd.concat([df_all, df])

        sess_df = df_all.drop([0],axis = 0)
        sess_df.dropna(axis=0, inplace=True)
        
        
    df_eda = pd.concat([df_eda, sess_df])
#df_eda.sort_values(by=['time'], ascending=True, inplace=True) # time 오름차순으로 정렬
#df_eda.drop(['time'], axis='columns', inplace=True)
```

```python
ibi = os.path.join(directory, 'IBI/')
files_ibi = os.listdir(ibi)
filescnt_ibi = [e for e in files_ibi]
df_ibi = pd.DataFrame()

for e in filescnt_ibi :
    ann = os.path.join(directory, 'IBI/', e)
    files = os.listdir(ann)
    filescnt_wav = [i for i in files if i.endswith('.csv')]
    df_all = pd.DataFrame()
    
    for i in range(0, len(filescnt_wav)) :
        file = filescnt_wav[i]
        df = pd.read_csv(os.path.join(ann, file), names=['order','IBI', 'time', 'id']) 
        df_all = pd.concat([df_all, df])

        sess_df = df_all.drop([0],axis = 0)
        sess_df.dropna(axis=0, inplace=True)
        
        
    df_ibi = pd.concat([df_ibi, sess_df])
#df_ibi.sort_values(by=['time'], ascending=True, inplace=True) # time 오름차순으로 정렬
#df_ibi.drop(['order','time'], axis='columns', inplace=True)
```

```python
temp = os.path.join(directory, 'TEMP/')
files_temp = os.listdir(temp)
filescnt_temp = [e for e in files_temp]
df_temp = pd.DataFrame()

for e in filescnt_temp :
    ann = os.path.join(directory, 'TEMP/', e)
    files = os.listdir(ann)
    filescnt_wav = [i for i in files if i.endswith('.csv')]
    df_all = pd.DataFrame()
    
    for i in range(0, len(filescnt_wav)) :
        file = filescnt_wav[i]
        df = pd.read_csv(os.path.join(ann, file), names=['TEMP', 'time', 'id']) 
        df_all = pd.concat([df_all, df])

        sess_df = df_all.drop([0],axis = 0)
        sess_df.dropna(axis=0, inplace=True)
        
        
    df_temp = pd.concat([df_temp, sess_df])
#df_temp.sort_values(by=['time'], ascending=True, inplace=True) # time 오름차순으로 정렬
#df_temp.drop(['time'], axis='columns', inplace=True)
```

## annotation 형식으로 EDA,IBI,TEMP 가공

##### EDA

```python
# 각 ID에 대한 첫번째와 마지막 EDA 값을 구함
start_eda = df_eda.groupby('id')['EDA'].first()
end_eda = df_eda.groupby('id')['EDA'].last()

# ID별로 첫번째와 마지막 EDA 값을 하나의 데이터프레임으로 만듬
df_segment = pd.concat([start_eda, end_eda], axis=1)
df_segment.columns = ['start', 'end']

# 중간에 해당하는 ID의 max EDA, min EDA, mean EDA 값을 구하여 데이터프레임에 추가
df_group = df_eda.groupby('id')['EDA'].agg(['max', 'min', 'mean']).reset_index()
df_segment = pd.merge(df_segment, df_group, on='id')

# Segment ID 컬럼 생성
df_segment['Segment ID'] = 'Sess01_script01_' + df_segment['id']

# 결과 출력
eda_seg = df_segment[['start', 'end', 'max', 'min', 'mean', 'Segment ID']]
eda_seg
```

##### IBI

###### order컬럼 삭제하고 IBI데이터가 문자열로 인식되기 때문에 숫자형으로 바꿈

```python
#df_ibi = df_ibi.iloc[:, 1:]

# to numeric
df_ibi['IBI'] = pd.to_numeric(df_ibi['IBI'], errors='coerce')
```

```python
# 각 ID에 대한 첫번째와 마지막 IBI 값을 구함
start_ibi = df_ibi.groupby('id')['IBI'].first()
end_ibi = df_ibi.groupby('id')['IBI'].last()

# ID별로 첫번째와 마지막 IBI 값을 하나의 데이터프레임으로 만듬
df_segment = pd.concat([start_ibi, end_ibi], axis=1)
df_segment.columns = ['start', 'end']

# 중간에 해당하는 ID의 max IBI, min IBI, mean IBI 값을 구하여 데이터프레임에 추가
df_group = df_ibi.groupby('id')['IBI'].agg(['max', 'min', 'mean']).reset_index()
df_segment = pd.merge(df_segment, df_group, on='id')

# Segment ID 컬럼 생성
df_segment['Segment ID'] = 'Sess01_script01_' + df_segment['id']

# 결과 출력
ibi_seg = df_segment[['start', 'end', 'max', 'min', 'mean', 'Segment ID']]
ibi_seg
```

##### TEMP

```python
# 각 ID에 대한 첫번째와 마지막 TEMP 값을 구함
start_temp = df_temp.groupby('id')['TEMP'].first()
end_temp = df_temp.groupby('id')['TEMP'].last()

# ID별로 첫번째와 마지막 TEMP 값을 하나의 데이터프레임으로 만듬
df_segment = pd.concat([start_temp, end_temp], axis=1)
df_segment.columns = ['start', 'end']

# 중간에 해당하는 ID의 max TEMP, min TEMP, mean TEMP 값을 구하여 데이터프레임에 추가
df_group = df_temp.groupby('id')['TEMP'].agg(['max', 'min', 'mean']).reset_index()
df_segment = pd.merge(df_segment, df_group, on='id')

# Segment ID 컬럼 생성
df_segment['Segment ID'] = 'Sess01_script01_' + df_segment['id']

# 결과 출력
temp_seg = df_segment[['start', 'end', 'max', 'min', 'mean', 'Segment ID']]
temp_seg
```
